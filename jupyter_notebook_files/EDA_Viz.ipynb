```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
#load data
dataframe = pd.read_csv('final.csv')
```

```
#data
dataframe.head()
```

```
#plot load values for a day
plt.plot(range(0, 48), dataframe['value'][:48])
plt.title('Daily Profile of Electricity Consumption')
plt.show()
```

```
#plot values for a month 48x30=1440
plt.plot(range(0, 1440), dataframe['value'][:1440])
plt.title('Monthly Profile of Electricity Consumption')
plt.show()
```

```
#plot values for a week 48x7=336
plt.plot(range(0, 336), dataframe['value'][:336])
plt.title('Weekly Profile of Electricity Consumption')
plt.show()
```
